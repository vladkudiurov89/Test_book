```
# Указываем путь
PATH = '/content/drive/My Drive/datasets/'
# Имя файла 
fn = 'транзакции.csv'
```

```
# Загружаем данные
trans = pd.read_csv(PATH + fn, header=None, names=['email', 'phone'])
# Указываем индекс с 1
trans.index +=1
# Выводим данные
display(trans)
```

```
# Создаём пустой словарь
logs = {}
# Указываем точку начала 
n = 1
# Запускаем цикл по данным с условиём
for tran in trans.itertuples():
    if tran[1] in logs and tran[2] in logs:
        logs[tran[1]] = min(logs[tran[1]], logs[tran[2]])
        logs[tran[2]] = min(logs[tran[1]], logs[tran[2]])
        continue
    
    elif tran[1] in logs and tran[2] not in logs:
        logs[tran[2]] = logs[tran[1]]
        continue
            
    elif tran[2] in logs and tran[1] not in logs:
        logs[tran[1]] = logs[tran[2]]
        continue
    
    else:
        logs[tran[1]] = n
        logs[tran[2]] = n
        n += 1
```

```
# Соединяем в один словарь
cnt = dict((k,0) for k in set(logs.values()))
for tran in trans.itertuples():
    if len(tran[1]):
        cnt[logs[tran[1]]] += 1
    else:
        cnt[logs[tran[2]]] += 1
```

```
# Получаем список уникальных значений из трансакций
value = cnt.values()
value
```
